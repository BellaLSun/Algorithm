### 分为两种情况，不分组进行排序和组内进行排序

- 如果只是单纯想对某一列进行排序，而不进行打序号的话直接使用.sort_values就可以了。下文是关于如何把序号也打上的

- #### 不分组排序

```python
import pandas as pd
a = pd.Series([7,8,9,8,9,9,4,10,7,9])
b = pd.Series([2,4,3,2,3,4,1,2,2,4])
c= pd.concat([a, b])
data = pd.DataFrame({'cate':a,'comment_num':b})
data
```

```python
data['sort_num']=data['comment_num'].rank(ascending=0,method='dense')
data
# 添加了一个新列，coment_num的排序先后，但是并没有在改变原表中数据的顺序
# dense的话，遇到一样的值，会给一样的排序
# method='first'，遇到一样的值，会按升序排序
```

```python
data['sort_num']=data['comment_num'].rank(ascending=1,method='dense')
data
# ascending=true，min=1.上边的是Max=1
```

```python
data['sort_num']=data['comment_num'].rank(ascending=0,method='min')
data
# method='min'，遇到一样的值，排序一样。但是下一个值会从真正的次序开始，而不是2开始。
```

```python
data['sort_num']=data['comment_num'].rank(ascending=0,method='max')
data
# method='min'，遇到一样的值，排序一样。但是这个值等于真正次序的最后一个值，而不是2开始。
```

### 组内排序

```python
a = pd.Series([7,8,9,8,9,9,4,10,7,9])
b = pd.Series([2,4,3,2,3,4,1,2,2,4])
c= pd.concat([a, b])
data = pd.DataFrame({'cate':a,'comment_num':b})
```

```python
data['group_sort']=data['comment_num'].groupby(data['cate']).rank(ascending=0,method='dense')
data
```

```python
a = pd.Series([7,8,9,8,9])
b = pd.Series([2,2,2,2,3])
c= pd.concat([a, b])
data = pd.DataFrame({'cate':a,'comment_num':b})
```

```python
data['group_sort']=data['comment_num'].groupby(data['cate']).rank(ascending=1,method='first')
data
```

好复杂，终于懂了。在comment_num中按照cate分组，比如cate=7是一组，cate=8是一组，然后按照comment_num的大小排序。

# 另一个案例

```python
import pandas as pd

df = pd.DataFrame([['A',1],['A',3],['A',2],['B',5],['B',9]], columns = ['name','score'])
df
```

**先排序，后分组**

先排序，后分组这种方法，基于一个前提，即分组不会改变组内的相对顺序。这种方法，首先采用sort_values的方法，对整个df进行排序，然后对相对字段进行group操作


```python
df.sort_values('score', ascending = False, inplace=True)
df
```

```python
res = df.groupby('name').head()
res
```

**先分组，后排序**

```python
df = pd.DataFrame([['A',1],['A',3],['A',2],['B',5],['B',9]], columns = ['name','score'])
df.groupby('name')
df
```

```python
df.sort_values('score',ascending=0,inplace=True)
df
```

**多列排序实现**

相当于直接先对name列排序，再对score列排序，这样也变相实现了分组排序的功能

```python
df = pd.DataFrame([['A',1],['A',3],['A',2],['B',5],['B',9]], columns = ['name','score'])
df.sort_values(['name','score'],ascending=[0,0],inplace=True)
df
```
